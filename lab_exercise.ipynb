```python
#Task 1: Import numpy Functions
import numpy as np
#Task 4: Add pyplot Functions to Our Portfolio
import matplotlib.pyplot as plt

#Task 2: Prepare Result Variable
data = [(1, 2, 3), (4, 5, 6), (7, 8, 9)]
netValue = np.zeros((len(data)))
```

```python
#Task 3: Sum Over Holdings' Values
for history in data:                        #This variable could technically be called something else, history is a little confusing
    print(history)                          #Prints (1, 2, 3)...
    netValue = np.add(netValue, history)    #The add function computes the addition of 2 arrays                     
print(netValue)  

#Task 5: First Look at The Portfolio's Value
plt.plot(netValue)
plt.show
```

```python
#Task 6: More Comprehensive Graph
for history in data:                        #This variable could technically be called something else, history is a little confusing
    print(history)                          #Prints (1, 2, 3)...
    netValue = np.add(netValue, history)    #The add function computes the addition of 2 arrays
    plt.plot(netValue)                      
print(netValue)                             #Prints [12. 15. 18]
plt.show
```

```python
#Task 7: Append to Portfolio.py
fig, axis = plt.subplots()
colors = ['#ff9999','#ee82ee','#66b3ff','#99ff99','#ffcc99']                                                    #Pretty colors for our pie chart
axis.pie(netValue, labels = data, autopct = '%1.1f%%', shadow = True, explode = (.2,0,0), colors = colors)
axis.set_aspect("equal")
```

```python
#Task 8: Example Flow Chart: Electricity Production
from matplotlib.sankey import Sankey
Sankey(flows=[0.32, 0.30, 0.20, 0.01, 0.17, -0.38, -0.37, -0.25], labels=['natural gas', 'coal', 'nuclear', 'petroleum', 'renewable', 'residential', 
'commercial', 'industrial'], orientations=[-1, -1, 0, -1, 1, -1, 0, 1]).finish()
plt.title("Sankey flow chart of U.S. electricity with default settings.")
plt.show()
```

```python
#Task 9: Frequency Distributions by Histogram
from random import randint
set1 = [0]*100
set2 = [0]*100
for x in range (100):
    set1[x] = randint(1,50)
    set2[x] = randint(1,50)

colors = ['#ff9999','#ee82ee','#66b3ff','#99ff99','#ffcc99'] 
plt.hist(set1, bins=25, alpha = .5, color = colors[0])                             #bin specifies the number of equal-width bins in the range
plt.hist(set2, bins=25, alpha = .5, color = colors[3])

plt.gca().set(title='Frequency Histogram', ylabel='Frequency')

```
